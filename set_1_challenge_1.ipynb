```python
import gwpy
from gwpy.timeseries import TimeSeries

#loading data
filename = "challenge1.gwf"
channel = "H1:CHALLENGE1"
data = TimeSeries.read(filename, channel)
sampling_rate = data.sample_rate.value
duration = data.duration.value
print(f"Sampling Rate: {sampling_rate}")
print(f"Duration: {duration}")

import warnings
warnings.filterwarnings("ignore", "Wswiglal-redir-stdio")
import lal
```

```python
#plotting data
plot = data.plot()
```

```python
#creating spectogram
specgram = data.spectrogram2(fftlength=4, overlap=2, window='hann') ** (1/2.)
plot = specgram.plot()
```

```python
gps = 2
qdata = data.q_transform(frange=(20, 200), qrange=(20, 30), outseg=(gps-5,gps+5))
plot = qdata.plot()
ax = plot.gca()
ax.set_epoch(gps)
ax.set_yscale('log')
ax.colorbar(label="Normalised energy")
plot.colorbars[0].mappable.set_clim(0,20)
plot.refresh()

```
